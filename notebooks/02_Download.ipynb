```python
import json
import sys
import yaml
from pathlib import Path
sys.path.append(str(Path("..").resolve()))
from src.config import load_config
import subprocess
```

```python
config = load_config('../config.local.yaml')
data_path = Path(config['paths']['bambi_data'])

with open('../data/flight_keys.json') as f:
    flight_keys = json.load(f)['Flight_keys']
```

```python
subprocess.run(["python", str(data_path/"download_from_zenodo.py"), '--help'], capture_output=True, text=True, cwd=str(data_path))
```

```python
test_flights = flight_keys 

result = subprocess.run(
    ["python", str(data_path / "download_from_zenodo.py"),
     "--flights"] + test_flights + [
     "--output-dir", str(Path("../data").resolve()), 
     "--unzip"
    ],
    capture_output=True,
    text=True,
    cwd=str(data_path), 
    encoding='utf-8',
    errors='replace'
)
print(result.stdout)
```
